```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Load the Iris dataset from a CSV file
iris_data = pd.read_csv("/kaggle/input/iris/Iris.csv")

# Print the first 5 rows of the data
iris_data.head()
```

```python
# number of (rows & columns) we have in the imported dataset
iris_data.shape
```

```python
# prints data type of each column
print(iris_data.info())
```

```python
# some statistal summary about each column

print(iris_data.describe())
```

```python
#total number of missiong values in each column:
print(iris_data.isnull().sum())
```

```python
# removing duplicate data from our data frame.
data = iris_data.drop_duplicates(subset ="Species",)
```

```python
sns.countplot(x='Species', data=iris_data, )
plt.show()
```

```python
# scatter plot
# grouped by specie types
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',
                hue='Species',  data=iris_data, )

plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',
                hue='Species', data=iris_data, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
sns.pairplot(iris_data.drop(['Id'], axis = 1),
             hue='Species', height=2)
```
